```python
import matplotlib.pyplot as plt
import numpy as np
x = ['1998', '2003', '2008', '2012']
y1 = np.array([10, 20, 10, 30])
y2 = np.array([20, 25, 15, 25])
y3 = np.array([12, 15, 19, 6])
y4 = np.array([10, 29, 13, 19])
plt.bar(x, y1, color='royalblue')
plt.bar(x, y2, bottom=y1, color='maroon')
plt.bar(x, y3, bottom=y1+y2, color='gray')
plt.bar(x, y4, bottom=y1+y2+y3, color='black')
plt.xlabel("Years")
plt.ylabel("Percentage in year")
plt.legend(["Additional Car", "First Car", "Shared Car", "Single Driver Car"])
plt.title("Proportion of cars by role of vehicle within households")
plt.show()
```
